```python
#Function to print uniques values in a columns
def print_unique(df):

    for column in df.columns:
        # Since we have all categorical data
        print(f"{column} column has {df[column].nunique()} values which are {df[column].unique()}")

print_unique(df)
```

```python
# Lets visualize that with a pie chart for every column

def plot_pie_chart(df, col):

    x= df[col].value_counts()
    #labels= df[col].unique()
    labels= df[col].value_counts().index
    plt.xlabel(col)
    plt.title(f'Pie Chart showing distribution of unique values in {col} column')
    plt.pie(x= x, labels= labels, autopct= '%1.1f%%')
    plt.show()

plot_pie_chart(df,'Sadness')
```

```python
for col in df.columns:

    plot_pie_chart(df,col)
```

### From the above charts we find out the distribution of values in specific columns
### Now column Suicidal Thoughts has two different values for YES, lets fix that

```python
df['Suicidal thoughts'].value_counts()
```

- As we can see there are two different values of YES, namely 'YES' and 'YES ' lets remove the the space from it

```python
df['Suicidal thoughts'] = list(map(lambda x: x.replace(" ","") ,df['Suicidal thoughts'].values))
```

```python
df['Suicidal thoughts'].value_counts()
```

```python
plot_pie_chart(df, 'Suicidal thoughts')
```
